#### Task 1(Pandas Series Basics)

```python
import numpy as np
import pandas as pd
```

```python
marks = [78,85,90,66,72]
marks_series = pd.Series(marks)
```

```python
for i in marks_series.values:
    print(i)
```

### TASK 3(Python Functionalities on Series)

```python
print(f"Maximum marks : {marks_series.max()}")
print(f"Minimum marks : {marks_series.min()}")
print(f"Sum of marks : {marks_series.sum()}")
print(f"Mean of marks : {marks_series.mean()}")
```

```python
lambda_pass = pd.Series(list((map(lambda x:x>70,marks_series.values))))
lambda_pass
#True : Passed
#False : Failed
```

```python
count = 0
for i in lambda_pass.values:
    if(i==True):
        count = count + 1
print(f"The no of students passed are {count}")
```

#### TASK 9(Mini Use Case : Sales Data Analysis)

```python
sales = {
    'Day' : ['Mon','Tue','Wed','Thurs','Fri'],
    'Revenue' : [1200,1500,900,2000,1800]
}

dr = pd.DataFrame(sales)
```

```python
print(f"Total Revenue is {dr['Revenue'].sum()}")
```

```python
print(f"Average Daily Revenue is {dr['Revenue'].mean()}")
```

```python
print(f"The Highest Revenue Day is : {((dr[dr['Revenue']==dr['Revenue'].max()])['Day'].values)}")
```

```python
print(f"The Days where Revenue>avearge is {list((dr[dr['Revenue']>dr['Revenue'].mean()])['Day'].values)}")
```

```python
dr.plot(x='Day',y='Revenue',kind='bar')
```
